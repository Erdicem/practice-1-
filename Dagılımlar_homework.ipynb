## Hisse Senedi DaÄŸÄ±lÄ±mÄ± 

```python
df=pd.read_csv('Google.csv')
```

```python
df.head()
```

```python
"""x-(x-1)/x-1
x= x zamanÄ±ndaki hisse senedi fiyatÄ±"""
```

```python
import pandas as pd 
ready=df.groupby('Date')['Close'].transform(lambda x: x - (x -1)/ x-1)
```

```python
plt.hist(ready)
plt.show()
```

```python
from scipy import stats

plt.figure(figsize=(8,5),dpi=100)
stats.probplot(ready,plot=plt)
plt.show()
```
