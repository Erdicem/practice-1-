## Çok Değiskenli analiz 

```python
import matplotlib.pyplot as plt 
import pandas as pd 
import numpy as np 
import seaborn as sns
import scipy.stats as stats
from scipy.stats.mstats import winsorize
import warnings

warnings.filterwarnings('ignore')
sns.set(style="whitegrid")
pd.options.display.float_format = '{:.3f}'.format
plt.rcParams['figure.dpi'] = 100
plt.rcParams['figure.figsize'] = (8,5.5)
```

```python
ogrenciler = pd.read_csv('StudentsPerformance.csv')
ogrenciler.columns = ['cinsiyet', 'koken', 'aile_egt_seviyesi', 'ogle_yemegi', 
                              'hazirlik_kursu', 'matematik_notu', 'okuma_notu', 'yazma_notu']
ogrenciler.head()
```

```python
plt.figure(figsize=(30,25))
listem=['matematik_notu','okuma_notu','yazma_notu']
listem_2=['cinsiyet','koken','aile_egt_seviyesi']
for i in range(3):
    plt.subplot(3,3,i+1)
    sns.barplot(ogrenciler[listem[i]],ogrenciler[listem_2[0]])
    plt.title(listem[i]+" "+listem_2[0])
    
for i in range(3):
    plt.subplot(3,3,i+4)
    sns.barplot(ogrenciler[listem[i]],ogrenciler[listem_2[1]])
    plt.title(listem[i]+" "+listem_2[1])
    
for i in range(3):
    plt.subplot(3,3,i+7)
    sns.barplot(ogrenciler[listem[i]],ogrenciler[listem_2[2]])
    plt.title(listem[i]+" "+listem_2[2])
    
plt.show()
```

```python
plt.figure(figsize=(18,8))
for i in range(3):
    plt.subplot(3,3,i+1)
    sns.barplot(ogrenciler['ogle_yemegi'],ogrenciler[listem[i]])
    plt.title("Ögle Yemegi &"+listem[i])
    
```

```python
plt.figure(figsize=(18,8))
for i in range(3):
    plt.subplot(3,3,i+1)
    sns.barplot(ogrenciler['hazirlik_kursu'],ogrenciler[listem[i]])
    plt.title("Hazırlık Kurus &"+listem[i])
```

```python
korelasyon_mat_ogrenciler = ogrenciler.corr()
display(korelasyon_mat_ogrenciler)
```

```python
plt.figure(figsize=(18,5))

sns.heatmap(korelasyon_mat_ogrenciler, square=True, annot=True, linewidths=.5, vmin=0, vmax=1, cmap='viridis')
plt.title("Korelasyon Matrisi (Öğrenci Notları)")

plt.show()
```
