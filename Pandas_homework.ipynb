```python
import pandas as pd
import numpy as np

df=pd.DataFrame(index=["a","b","c","d","e","f","g","h","i","j"])
df.index.name='İndex'
df
```

```python
def Asalmisin (sayi):
    if sayi >1:
        for i in range(2,sayi):
            if sayi%i==0:
                return False
            return True
    else:
        return False
```

```python
df['Asal_Sayilar'] =[x for x in range(2, 30)
     if all(x % y != 0 for y in range(2, x))]
df
```

```python
def fibonacci(n):
    fibo=[0,1]
    [fibo.append(fibo[x-2]+fibo[x-1]) for x in range(2,n)]
    return fibo
```

```python
fibonacci(10)
```

```python
df['Fibo_Sayi']=fibonacci(10)
df
```

```python
import pandas as pd
import seaborn as sns
df = sns.load_dataset('titanic')
df.head()
```

```python
# her cinsiyetin ort. yaşı 
df.groupby('sex')['age'].mean()
```

```python
# 30 yaşından küçük olanların kurtulma oranı
age =pd.cut(df['age'],[0,30,90])
df.pivot_table('survived', index=[age,'sex'] )
```

```python
df.pivot_table('survived', ['sex',age], 'class')
```

```python
df.groupby(['sex',age,'class'])['survived'].mean().unstack()
```
